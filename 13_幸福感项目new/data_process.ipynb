## 数据处理的思路：
  - 0、先删除一些缺省比较多的记录，或者特征？（一般在训练阶段没有标签的数据我们也是不能要的）
  - 1、时间的数据需要进行时间数据的提取 time striptime()
  - 2、连续的数据进行分段处理===》转换为离散数据
  - 3、文本数据的提取：oneHOT，TFIDF，词向量  （文本向量化）
  - 4、对连续型的数据进行缺失值的填充 一般可以填充为0，均值，众数，中位数，经验数据
  - 5、对离散数据进行缺失值的填充 (缺失值单独作为一个类别，根据标签类别数量进行填充，经验数据) 哑编码
  - 6、对于我们处理好后的数据，还可以进行一个相关性的检验，对于一些相关系数绝对值比较大的特征只保留其中一个特征（可以考虑的，但是不是说效果就一定好，这样做好处是特征降维了，节约我们的训练成本）
  - 7、根据已有特征提取新的特征。
  # PCA，标准化，归一化、特征选择等特征工程这些工作是在我们把数据进行预处理后（数值型的特征）在进行的

```python
### 在统计nan值前对数据中的异常值进行替换成nan值
## -1 = 不适用; -2 = 不知道; -3 = 拒绝回答; -8 = 无法回答;
### inplace=True 表示在原对象上进行我们的修改
### inplace=False 表示我们修改后会返回一个新对象，原对象不会改变，所以这时候我们要重新赋值
all_datas.replace([-1,-2,-3,-8],np.nan,inplace=True)
print(all_datas.info(verbose=True,null_counts=True))
```

```python
# todo：其实 可以对数据进行时间的提取 年份，时间段 time striptime()
# print(time.strptime('2015/8/4 14:18','%Y/%m/%d %H:%M'))
### note：todo：把其他的时间数据也保留下来
def time_strip(str):
    '''
    对时间数据进行提取保留year和hour
    :param str:
    :return:
    '''
    # print(str[0])
    date = time.strptime(str[0],'%Y/%m/%d %H:%M')
    return (date.tm_year,date.tm_hour,date.tm_wday)
# print(all_datas['survey_time'].value_counts())
all_datas[['survey_year','survey_hour','survey_week_day']] = all_datas[['survey_time']].apply(lambda x:pd.Series(time_strip(x)),axis=1)
print(all_datas[['survey_year','survey_hour','survey_week_day']].head())
### 后面 'survey_time' 不再需要了我们就删除掉
all_datas.drop(['survey_time'],axis=1,inplace=True)
```

```python
## todo:'income'，'inc_exp' 做一个字段：收入是否达到期望收入
### 判断一下收入是否符合预期
def if_exp_inc(x):
    income, inc_exp = x[0],x[1]
    if income>=inc_exp:
        return 1
    else:
        return 0
all_datas['if_exp_inc'] = all_datas[['income','inc_exp']].apply(if_exp_inc,axis=1)
print(all_datas['if_exp_inc'].value_counts(dropna=False))
```
